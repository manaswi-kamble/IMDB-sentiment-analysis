```python
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing.text import one_hot
```

```python
from tensorflow.keras.preprocessing.text import one_hot, Tokenizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dropout, Dense, Flatten, GlobalMaxPooling1D, Embedding, Conv1D, LSTM
from sklearn.model_selection import train_test_split
from keras.preprocessing.sequence import pad_sequences
```

```python
df = pd.read_csv(path, nrows=100)
```

```python
df.head(10)
```

```python
df['sentiment'].replace({"positive":1, "negative":0}, inplace=True)
df.head(10)
```

```python
X = np.array(df['review'].values)
y = np.array(df['sentiment'].values)
```

```python
X_filtered = []
for review in X:

  #for lower casing
  review = review.lower()

  #for removing punctuations
  for i in review:
    punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    if i in punc:
      review = review.replace(i, " ")

  X_filtered.append(review)

print(X_filtered)
```

```python
#converting sentences to vectors

vocabulary_Size = 5000

one_hot_encoded = [one_hot(review, vocabulary_Size) for review in X_filtered]

print(one_hot_encoded)
```

```python
#padding senetences

max_length = 500

X_padded = pad_sequences(one_hot_encoded, max_length, padding='post')

```

```python

X_train,X_test,y_train,y_test = train_test_split(X_padded,y,test_size=0.2)
```

embedding layer transforms our vector.
here each word in our dataset is now represented by a single vector.

 but the size of this entire vector is very huge.. which is basically an efficiency. so, in order to represent one word we have to create a vector of 5000 numbers.

 so in order to reduce the size of this vector and to add some semantic meaning we use the embedding layer

 embedded vector basically takes two parameters
 1. vocabulary size
 2. embedded vector size
 (how much we want to reduce the original vector size)

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Embedding

# Define vocabulary size
vocabulary_size = 5000  # Example vocabulary size

embeded_vector_size = 35
model = Sequential()
model.add(Embedding(vocabulary_size, embeded_vector_size, input_length=max_length))
model.add(LSTM(100))
model.add(Dense(1, activation="sigmoid"))

print(model.summary())
print("Model Creation Completed!")
```

```python
model.compile(loss="binary_crossentropy", metrics=["accuracy"])
model.fit(X_train, y_train, epochs=10)
```

```python
score = model.evaluate(X_test, y_test)
print("Accuracy: ", score)
```

```python
model.save("IMDB_sentiment_analysis.keras")
```
